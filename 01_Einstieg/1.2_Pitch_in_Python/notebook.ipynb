# 1.2 — Drawing a Football Pitch in Python from Scratch

We build a reusable `draw_pitch()` function step by step using only matplotlib.

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc

FIGURES = 'figures'
```

---
## Step 1: Outer Boundary

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.set_facecolor('white')

ax.add_patch(plt.Rectangle((0, 0), 120, 80,
             fill=False, edgecolor='#333333', linewidth=2))

ax.set_xlim(-3, 123)
ax.set_ylim(-3, 83)
ax.set_aspect('equal')
ax.axis('off')
plt.savefig(f'{FIGURES}/build_01_boundary.png', dpi=150, bbox_inches='tight')
plt.show()
```

---
## Step 2: Halfway Line and Centre Circle

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.set_facecolor('white')
lc, lw = '#333333', 1.5

ax.add_patch(plt.Rectangle((0, 0), 120, 80, fill=False, edgecolor=lc, linewidth=lw))

# Halfway line
ax.plot([60, 60], [0, 80], color=lc, linewidth=lw)

# Centre circle (radius = 10)
ax.add_patch(plt.Circle((60, 40), 10, fill=False, edgecolor=lc, linewidth=lw))
ax.plot(60, 40, 'o', color=lc, markersize=3)

ax.set_xlim(-3, 123); ax.set_ylim(-3, 83)
ax.set_aspect('equal'); ax.axis('off')
plt.savefig(f'{FIGURES}/build_02_centre.png', dpi=150, bbox_inches='tight')
plt.show()
```

---
## Step 3: Penalty Areas and 6-Yard Boxes

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.set_facecolor('white')
lc, lw = '#333333', 1.5

ax.add_patch(plt.Rectangle((0, 0), 120, 80, fill=False, edgecolor=lc, linewidth=lw))
ax.plot([60, 60], [0, 80], color=lc, linewidth=lw)
ax.add_patch(plt.Circle((60, 40), 10, fill=False, edgecolor=lc, linewidth=lw))
ax.plot(60, 40, 'o', color=lc, markersize=3)

# Penalty areas
ax.add_patch(plt.Rectangle((0, 18), 18, 44, fill=False, edgecolor=lc, linewidth=lw))
ax.add_patch(plt.Rectangle((102, 18), 18, 44, fill=False, edgecolor=lc, linewidth=lw))

# 6-yard boxes
ax.add_patch(plt.Rectangle((0, 30), 6, 20, fill=False, edgecolor=lc, linewidth=lw))
ax.add_patch(plt.Rectangle((114, 30), 6, 20, fill=False, edgecolor=lc, linewidth=lw))

ax.set_xlim(-3, 123); ax.set_ylim(-3, 83)
ax.set_aspect('equal'); ax.axis('off')
plt.savefig(f'{FIGURES}/build_03_penalty.png', dpi=150, bbox_inches='tight')
plt.show()
```

---
## Step 4: Goals, Penalty Spots and Arcs — Complete Pitch

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.set_facecolor('white')
lc, lw = '#333333', 1.5

ax.add_patch(plt.Rectangle((0, 0), 120, 80, fill=False, edgecolor=lc, linewidth=lw))
ax.plot([60, 60], [0, 80], color=lc, linewidth=lw)
ax.add_patch(plt.Circle((60, 40), 10, fill=False, edgecolor=lc, linewidth=lw))
ax.plot(60, 40, 'o', color=lc, markersize=2)
ax.add_patch(plt.Rectangle((0, 18), 18, 44, fill=False, edgecolor=lc, linewidth=lw))
ax.add_patch(plt.Rectangle((102, 18), 18, 44, fill=False, edgecolor=lc, linewidth=lw))
ax.add_patch(plt.Rectangle((0, 30), 6, 20, fill=False, edgecolor=lc, linewidth=lw))
ax.add_patch(plt.Rectangle((114, 30), 6, 20, fill=False, edgecolor=lc, linewidth=lw))

# Goals (dashed, extend outside pitch)
ax.add_patch(plt.Rectangle((-2, 36), 2, 8, fill=False, edgecolor=lc, linewidth=lw, linestyle='--'))
ax.add_patch(plt.Rectangle((120, 36), 2, 8, fill=False, edgecolor=lc, linewidth=lw, linestyle='--'))

# Penalty spots
ax.plot(12, 40, 'o', color=lc, markersize=3)
ax.plot(108, 40, 'o', color=lc, markersize=3)

# Penalty arcs (partial circles outside the penalty area)
ax.add_patch(Arc((12, 40), 20, 20, angle=0, theta1=307, theta2=233, color=lc, linewidth=lw))
ax.add_patch(Arc((108, 40), 20, 20, angle=0, theta1=127, theta2=53, color=lc, linewidth=lw))

ax.set_xlim(-3, 123); ax.set_ylim(-3, 83)
ax.set_aspect('equal'); ax.axis('off')
plt.savefig(f'{FIGURES}/build_04_complete.png', dpi=150, bbox_inches='tight')
plt.show()
```

---
## The Reusable Function

```python
def draw_pitch(ax, pitch_color='white', line_color='#333333'):
    lc, lw = line_color, 1.5
    ax.set_facecolor(pitch_color)
    for r in [
        plt.Rectangle((0,0),120,80,fill=True,facecolor=pitch_color,edgecolor=lc,linewidth=lw),
        plt.Rectangle((0,18),18,44,fill=False,edgecolor=lc,linewidth=lw),
        plt.Rectangle((102,18),18,44,fill=False,edgecolor=lc,linewidth=lw),
        plt.Rectangle((0,30),6,20,fill=False,edgecolor=lc,linewidth=lw),
        plt.Rectangle((114,30),6,20,fill=False,edgecolor=lc,linewidth=lw),
        plt.Rectangle((-2,36),2,8,fill=False,edgecolor=lc,linewidth=lw,linestyle='--'),
        plt.Rectangle((120,36),2,8,fill=False,edgecolor=lc,linewidth=lw,linestyle='--'),
    ]:
        ax.add_patch(r)
    ax.plot([60,60],[0,80],color=lc,linewidth=lw)
    ax.add_patch(plt.Circle((60,40),10,fill=False,edgecolor=lc,linewidth=lw))
    ax.plot(60,40,'o',color=lc,markersize=2)
    ax.plot(12,40,'o',color=lc,markersize=3)
    ax.plot(108,40,'o',color=lc,markersize=3)
    ax.add_patch(Arc((12,40),20,20,angle=0,theta1=307,theta2=233,color=lc,linewidth=lw))
    ax.add_patch(Arc((108,40),20,20,angle=0,theta1=127,theta2=53,color=lc,linewidth=lw))
    ax.set_xlim(-3,123); ax.set_ylim(-3,83)
    ax.set_aspect('equal'); ax.axis('off')
```

---
## Light vs Dark Theme

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))
fig.patch.set_facecolor('#dddddd')

draw_pitch(ax1, pitch_color='white', line_color='#333333')
ax1.set_title('Light', fontsize=14, fontweight='bold', color='#222', pad=12)

draw_pitch(ax2, pitch_color='#2d6a2d', line_color='white')
ax2.set_title('Dark Green', fontsize=14, fontweight='bold', color='#222', pad=12)

plt.tight_layout(pad=2)
plt.savefig(f'{FIGURES}/pitch_both_themes.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Bonus: mplsoccer

If you don't want to build the pitch manually every time, mplsoccer gives you one in two lines.

```python
# pip install mplsoccer
from mplsoccer import Pitch

pitch = Pitch(pitch_color='grass', line_color='white', stripe=True)
fig, ax = pitch.draw(figsize=(12, 8))
plt.show()
```
